## Data Preprocessing

```python
from googletrans.client import Translated

## language Detect
def lang_detect(data):
    lang = detect(data)
    return lang

# review_translate
def lang_translator(data):
    translor = Translator()
    translated_text = translor.translate(data)
    return translated_text.text


## Removing Spaces
def remove_spaces(data):
    clean_text = data.replace("\\n", " ").replace('\t',' ').replace('\\', ' ')
    return clean_text



# expand text
def expand(data):
    expand_text = contractions.fix(data)
    return expand_text



# handling accented characters
def handling_accented(data):
    fixed_text = unidecode(data)
    return fixed_text



# removing stopwords
stopword_list = stopwords.words()
stopword_list.remove("no")
stopword_list.remove("nor")
stopword_list.remove("not")



# ceanining text
def clean_data(data):
    tokens = word_tokenize(data)
    clean_text = [word.lower() for word in tokens if (word not in punctuation) and (word.lower() not in stopword_list) and (len(word)>2) and (word.isalpha())]
    return clean_text



# autoCorrecting word
def autocorrect(data):
    spell = autocorrect.Speller(lang="en")
    corrected_text = spell(data)
    return corrected_text



# Lammitization
def lemmatization(data):
    lemmatizer = WordNetLemmatizer()
    final_data = []
    for word in data :
        lemmatized_word = lemmatizer.lemmatize(word)
        final_data.append(lemmatized_word)
    return " ".join(final_data)
```

### 1. Unigram

```python
def ngram_extractor(data, ngram_range):
    tokens = word_tokenize(data)
    ngram = ngrams(tokens, ngram_range)
    ngram_list = []
    for word in ngram:
        ngram_list.append(' '.join(word))
    return ngram_list


# Unigram list
unigram_list = df.Emails.apply(lambda x : ngram_extractor(x,1))

final_grams = []
for gram in unigram_list:
    final_grams.extend(gram)

most_common = Counter(final_grams).most_common(25)
print(most_common)
```

### 2. Bigram

```python
## Bigram list
Bigram_list = df.Emails.apply(lambda x : ngram_extractor(x,2))

final_bigrams = []
for gram in Bigram_list:
    final_bigrams.extend(gram)

most_two = Counter(final_bigrams).most_common(25)
print(most_two)
```

## WordCloud
- topic which is talk about most is highlighted in Biger font and the topic which are talk less is highlighted in Small fonts

```python
path = "/Internship/Task 4 Email Spam Detection/Email_spam Project/data"
def wordcloud(data, column, path=path):
    df_ = data[column].str.cat(sep=' ')
    text = ' '.join([word for word in df_.split()])
    wordcloud = WordCloud(scale=3,
                          width=700,
                          height=500,
                          background_color='black', 
                          collocations=True).generate(text)
    # saving model
    wordcloud.to_file(f"{path}/wordcloud.png")
    
    plt.figure(figsize=(10,8))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    
wordcloud(df,'Emails')
```

#### Yet Another Keyword Extractor (Yake)
- It is a lightweight, unsupervised automatic keyword extraction method that relies on statistical text features extracted from individual documents to identify the most relevant keywords in the text
- It is interesting that YAKE! pays attention to capital letters and gives more importance to words that start with a capital letter.

```python
def yake_extractor(data):
    extractor = yake.KeywordExtractor()
    keywords = extractor.extract_keywords(data)
    yake_kw_list = []
    for kw in keywords:
        yake_kw_list.append(kw[0])

    return yake_kw_list

keywords = df.Emails.apply(yake_extractor)
```

```python
yake_word_extracted = []
for word in keywords:
    yake_word_extracted.extend(word)

yake_count = Counter(yake_word_extracted).most_common(1000)
yake_count
```

#### Rapid Automatic Keyword Extraction
- Rake is short for Rapid Automatic Keyword Extraction and it is a method of extracting keywords from individual documents. It can also be applied to new fields very easily and is very effective in dealing with multiple types of documents, especially text that requires specific grammatical conventions. Rake identifies key phrases in a text by analyzing the occurrence of a word and its compatibility with other words in the text (co-occurrence).

```python
def rake_extractor(data):
    keyword_extractor = Rake()
    keyword_extractor.extract_keywords_from_text(data)
    return keyword_extractor.get_ranked_phrases()


rake_extractor = df.Emails.apply(rake_extractor)
```

```python
def vectorizer(list_of_doc, model):
    features = []
    
    # generate a vector of size vecotrs_size
    for rew in list_of_doc:
        zero_vect = np.zeros(model.vector_size)
        
    # append the word vectores in vectors list
    vectors = []
    for word in rew:
        if word in model.wv:
            try:
                vectors.append(model.wv[word])
            except:
                continue
    if vectors:
        vectors = np.asarray(vectors)
        avg_vectors = vectors.mean(axis=0)
        features.append(avg_vectors)
    else:
        features.append(zero_vect)

    return vectors
```
